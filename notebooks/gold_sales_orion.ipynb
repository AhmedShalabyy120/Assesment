# Preparing the gold layer for the sales

```python
import pandas as pd
```

```python
df_spark = spark.table("silver.sales")
```

```python
df_spark.printSchema()
```

```python
df=df_spark.toPandas()
```

```python
df.info()
```

# Normalization 

```python
# Creating dim_customers

dim_customers = df[['Customer_Code','CustomerKey','Name','Education','Occupation' ]].drop_duplicates()


```

```python
dim_customers.head()
```

```python
# Creating dim_geography


dim_geography= df[['City', 'State', 'CountryRegion', 'Continent']].drop_duplicates().reset_index(drop=True)
dim_geography['GeographyKey'] = dim_geography.index + 1
```

```python
dim_geography.head()
```

```python
# Creating dim_product

dim_product = df[['ProductKey', 'Product_Name', 'Brand', 'Category', 'Subcategory']].drop_duplicates()
```

```python
dim_product.head()
```

```python
df = df.merge(dim_geography, on=['City', 'State', 'CountryRegion', 'Continent'], how='left')
```

```python
df.shape
```

```python
df.head()
```

```python
fact_sales= df[['CustomerKey', 'ProductKey', 'GeographyKey', 
                 'OrderDate', 'Quantity', 'Net_Price', 'TotalRevenue']]
```

```python
fact_sales['DateKey'] = pd.to_datetime(
    fact_sales['OrderDate'],
    errors='coerce'
).dt.strftime('%Y%m%d')
```

```python
fact_sales['DateKey'] = fact_sales['DateKey'].astype(int)
```

```python
fact_sales.head()
```

# Saving the Facts and Dims in the gold layer

```python
spark_fact_sales = spark.createDataFrame(fact_sales)
spark_dim_geography = spark.createDataFrame(dim_geography)
spark_dim_product = spark.createDataFrame(dim_product)
spark_dim_customers = spark.createDataFrame(dim_customers)
```

```python
spark_fact_sales.write.mode('overwrite').format("delta").saveAsTable("gold.fact_sales")
spark_dim_geography.write.mode('overwrite').format("delta").saveAsTable("gold.dim_geography")
spark_dim_product.write.mode('overwrite').format("delta").saveAsTable("gold.dim_product")
spark_dim_customers.write.mode('overwrite').format("delta").saveAsTable("gold.dim_customers")
```
