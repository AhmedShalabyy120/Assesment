# Preparing the silver layer for forecast

```python
df_spark = spark.table("bronze.forecast")
```

```python
df = df_spark.toPandas()
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df = df.apply(lambda col: col.str.strip() if col.dtype == 'object' else col)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape 
```

```python
df_spark_gold = spark.createDataFrame(df)
```

```python
df_spark_gold.write.mode("overwrite").format("delta").saveAsTable("gold.forecast")
```
