```python
# Removing extra spaces from the attributes of each column

df = df.apply(lambda col: col.str.strip() if col.dtype == 'object' else col)
```

```python
# Removing  CS to all Customer Codes
df['Customer_Code'] = df['Customer_Code'].str.replace('CS', '')
```

```python
df[df['Customer_Code'].str.startswith('CS')]
```

```python
df['Customer_Code'] = df['Customer_Code'].astype(int)
```

```python
df.dtypes
```

```python
# Round Net Price to 2 decimals 
df["Net_Price"]=round(df["Net_Price"],2)
```

# Filling Missing Data

```python
df['Education'] = df['Education'].fillna("Unknown")
```

```python
df['Occupation'] = df['Occupation'].fillna("Unknown")
```

```python
# Filling the missing name with Customer_and his code for example Customer_1111

mask = df['Name'].isna()

df.loc[mask, 'Name'] = (
    'Customer_' +
    df.loc[mask, 'Customer_Code'].astype(str)
)
```

```python
df.groupby('Name')['TotalRevenue'].sum().reset_index().sort_values('TotalRevenue', ascending=False)
```

# Feature Engineering

```python
df['TotalRevenue'] = df['Net_Price'] * df['Quantity']
```

# Saving the table in the silver Layer

```python
df_spark_silver = spark.createDataFrame(df)
```

```python
df_spark_silver.write.mode('overwrite').format("delta").saveAsTable("silver.sales")
```
