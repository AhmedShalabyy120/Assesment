# Creating dim_date

```python
import pandas as pd

dates = pd.date_range(start='2008-01-01', end='2009-12-31', freq='D')

dim_date = pd.DataFrame({
    'DateKey':  dates.strftime('%Y%m%d').astype(int),  # 20080101
    'Date':     dates,
    'Year':     dates.year,
    'Month':    dates.month,
    'Quarter':  dates.quarter,
    'MonthName': dates.strftime('%B'),
    'Weekday':  dates.strftime('%A')
})
```

```python
dim_date.head()
```

```python
spark_dim_date = spark.createDataFrame(dim_date)
```

```python
spark_dim_date.write.mode('overwrite').format('delta').saveAsTable("gold.dim_date")
```
