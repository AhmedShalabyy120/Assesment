# Preparing the bronze layer

```python
sales_path = "abfss://f58da772-84ec-4cd9-9bcb-21a5ba1f0cb5@onelake.dfs.fabric.microsoft.com/0675e500-0759-4be8-b043-65e62c1416ce/Files/bronze/Sales.json"

df_sales = (
    spark.read
    .option("multiLine", "true")
    .option("mode", "PERMISSIVE")
    .json(sales_path)
)

```

```python
df_sales.printSchema()
```

```python
display(df_sales.limit(5))
```

```python
forecast_path ="abfss://f58da772-84ec-4cd9-9bcb-21a5ba1f0cb5@onelake.dfs.fabric.microsoft.com/0675e500-0759-4be8-b043-65e62c1416ce/Files/bronze/forecast.json"
df_forecast = (
    spark.read
    .option("multiLine", "true")
    .option("mode", "PERMISSIVE")
    .json(forecast_path)
)
```

```python
df_forecast.printSchema()
```

```python
display(df_forecast.limit(5))
```

```python
# In order to save the file as delta table some columns contain spaces so we must replace them with underscore

df_sales = df_sales.toDF(
    *[c.replace(" ", "_") for c in df_sales.columns]
)
```

```python
df_sales.write.mode('overwrite').format('delta').saveAsTable('bronze.sales')
```

```python
df_forecast.write.mode('overwrite').format('delta').saveAsTable('bronze.forecast')
```
